```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("training.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.drop('Unnamed: 133', axis = 1)
```

```python
df.isnull().any()
```

```python
df['prognosis'].value_counts()
```

```python
df2 = pd.read_csv('Testing.csv')
```

```python
df2.head()
```

```python
X_train = df.drop('prognosis', axis =1)
y_train = df['prognosis']
X_test = df2.drop('prognosis', axis =1)
y_test = df2['prognosis']
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# Traning random forest model
mod1 = RandomForestClassifier(n_estimators = 100,n_jobs = 5, criterion= 'entropy',random_state = 42)
mod1.fit(X_train,y_train.values.ravel())
```

```python
y_pred = mod1.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import classification_report
```

```python
report = classification_report(y_test, y_pred, output_dict=True)
pd.DataFrame(report).transpose()
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test, y_pred)
pd.DataFrame(cm)
```
